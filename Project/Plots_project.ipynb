## Gestione moduli e caricamento dati

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

n_data = pd.read_csv("Results_n.csv")  
n_data  
```

```python
m_data = pd.read_csv("Results_m.csv")
m_data
```

## Analisi per n variabile 

### Grafici individuali (scala lineare)

```python
fig, axes = plt.subplots(2, 2, figsize = (13, 8))

fig.suptitle("Algoritmi individuali (scala lineare)")

axes[0][0].plot(n_data["n"], n_data["CountingSort"], color = "#e60049", label = "CountingSort")

axes[0][0].set_title("CountingSort")
axes[0][0].set_xlabel("size (n)")
axes[0][0].set_ylabel("time (ns)")

axes[0][1].plot(n_data["n"], n_data["QuickSort"], color = "#0bb4ff", label = "QuickSort")

axes[0][1].set_title("QuickSort")
axes[0][1].set_xlabel("size (n)")
axes[0][1].set_ylabel("time (ns)")

axes[1][0].plot(n_data["n"], n_data["InsertionSort"], color = "#87bc45", label = "InsertionSort")

axes[1][0].set_title("InsertionSort")
axes[1][0].set_xlabel("size (n)")
axes[1][0].set_ylabel("time (ns)")

axes[1][1].plot(n_data["n"], n_data["QuickSort3Ways"], color = "#edbf33", label = "QuickSort-3-Ways")

axes[1][1].set_title("QuickSort-3-Ways")
axes[1][1].set_xlabel("size (n)")
axes[1][1].set_ylabel("time (ns)")

handles, labels = axes[0][0].get_legend_handles_labels()

fig.tight_layout(rect = [0, 0.15, 1, 1])

plt.show()
```

### Grafici individuali (scala logaritmica)

```python
fig, axes = plt.subplots(2, 2, figsize = (13, 8))

fig.suptitle("Algoritmi individuali (scala logaritmica)")

axes[0][0].plot(n_data["n"], n_data["CountingSort"], color = "#e60049", label = "CountingSort")

axes[0][0].set_title("CountingSort")
axes[0][0].set_xlabel("size (n)")
axes[0][0].set_ylabel("time (ns)")
axes[0][0].set_xscale("log")
axes[0][0].set_yscale("log")

axes[0][1].plot(n_data["n"], n_data["QuickSort"], color = "#0bb4ff", label = "QuickSort")

axes[0][1].set_title("QuickSort")
axes[0][1].set_xlabel("size (n)")
axes[0][1].set_ylabel("time (ns)")
axes[0][1].set_xscale("log")
axes[0][1].set_yscale("log")

axes[1][0].plot(n_data["n"], n_data["InsertionSort"], color = "#87bc45", label = "InsertionSort")

axes[1][0].set_title("InsertionSort")
axes[1][0].set_xlabel("size (n)")
axes[1][0].set_ylabel("time (ns)")
axes[1][0].set_xscale("log")
axes[1][0].set_yscale("log")

axes[1][1].plot(n_data["n"], n_data["QuickSort3Ways"], color = "#edbf33", label = "QuickSort-3-Ways")

axes[1][1].set_title("QuickSort-3-Ways")
axes[1][1].set_xlabel("size (n)")
axes[1][1].set_ylabel("time (ns)")
axes[1][1].set_xscale("log")
axes[1][1].set_yscale("log")

handles, labels = axes[0][0].get_legend_handles_labels()

fig.tight_layout(rect = [0, 0.15, 1, 1])

plt.show()
```

### Grafici comparativi

```python
fig, axes = plt.subplots(1, 2, figsize = (13, 6))

fig.suptitle("Confronto algoritmi di ordinamento")

axes[0].plot(n_data["n"], n_data["CountingSort"], color = "#e60049", label = "CountingSort")
axes[0].plot(n_data["n"], n_data["QuickSort"], color = "#0bb4ff", label = "QuickSort")
axes[0].plot(n_data["n"], n_data["InsertionSort"], color = "#87bc45", label = "InsertionSort")
axes[0].plot(n_data["n"], n_data["QuickSort3Ways"], color = "#edbf33", label = "QuickSort-3-Ways")

axes[0].set_title("Scala lineare")

axes[0].set_xlabel("size (n)")
axes[0].set_ylabel("time (ns)")

axes[1].plot(n_data["n"], n_data["CountingSort"], color = "#e60049")
axes[1].plot(n_data["n"], n_data["QuickSort"], color = "#0bb4ff")
axes[1].plot(n_data["n"], n_data["InsertionSort"], color = "#87bc45")
axes[1].plot(n_data["n"], n_data["QuickSort3Ways"], color = "#edbf33")

axes[1].set_title("Scala logaritmica")

axes[1].set_xlabel("size (n) [log]")
axes[1].set_ylabel("time (ns) [log]")

axes[1].set_xscale("log")
axes[1].set_yscale("log")

handles, labels = axes[0].get_legend_handles_labels()

fig.tight_layout(rect = [0, 0.15, 1, 1])
fig.legend(handles, labels, loc = 'lower center', ncol = 2) 

plt.show()
```

### Standard deviation

```python
fig, axes = plt.subplots(2, 2, figsize = (13, 8))

fig.suptitle("Standard deviation - n variabile")

axes[0][0].fill_between(n_data["n"], n_data["CountingSort"] - n_data["sd_CS"], n_data["CountingSort"] + n_data["sd_CS"], color = "#e60049", alpha = 0.6, label = "CountingSort")

axes[0][0].set_title("CountingSort")
axes[0][0].set_xlabel("size (n)")
axes[0][0].set_ylabel("time (ns)")
axes[0][0].set_xscale("log")
axes[0][0].set_yscale("log")

axes[0][1].fill_between(n_data["n"], n_data["QuickSort"] - n_data["sd_QS"], n_data["QuickSort"] + n_data["sd_QS"], color = "#0bb4ff", alpha = 0.6, label = "QuickSort")

axes[0][1].set_title("QuickSort")
axes[0][1].set_xlabel("size (n)")
axes[0][1].set_ylabel("time (ns)")
axes[0][1].set_xscale("log")
axes[0][1].set_yscale("log")

axes[1][0].fill_between(n_data["n"], n_data["InsertionSort"] - n_data["sd_IS"], n_data["InsertionSort"] + n_data["sd_IS"], color = "#87bc45", alpha = 0.6, label = "InsertionSort")

axes[1][0].set_title("InsertionSort")
axes[1][0].set_xlabel("size (n)")
axes[1][0].set_ylabel("time (ns)")
axes[1][0].set_xscale("log")
axes[1][0].set_yscale("log")

axes[1][1].fill_between(n_data["n"], n_data["QuickSort3Ways"] - n_data["sd_QS3W"], n_data["QuickSort3Ways"] + n_data["sd_QS3W"], color = "#edbf33", alpha = 0.6, label = "QuickSort3Ways")

axes[1][1].set_title("QuickSort-3-Ways")
axes[1][1].set_xlabel("size (n)")
axes[1][1].set_ylabel("time (ns)")
axes[1][1].set_xscale("log")
axes[1][1].set_yscale("log")

handles, labels = axes[0][0].get_legend_handles_labels()

fig.tight_layout(rect = [0, 0.15, 1, 1])

plt.show()
```
